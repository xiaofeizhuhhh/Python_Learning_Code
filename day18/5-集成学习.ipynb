```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
x=np.arange(0,12).reshape(3,4)
y=np.array([True,False,True])
print(y)
print(x)
x[y,1]#y的0号和2号元素均为True，这里是取x的第0行和第2行
```

```python
#x是特征，y是标签
"""
make_moons()函数生成一个用于二分类或聚类测试的经典数据集——双月形数据集
n_samples=50000：生成 50,000 个样本点。
noise=0.3：为数据添加标准差为 0.3 的高斯噪声，使两个半月形状重叠程度增加，增加分类难度。
random_state=42：固定随机种子，确保每次运行生成的噪声和样本分布一致，便于复现实验。
返回的 x 是一个形状为 (50000, 2) 的数组，每行是一个样本的两个特征（横纵坐标）。
返回的 y 是一个形状为 (50000,) 的数组，每个值是样本的类别标签（0 或 1），分别对应两个不同的半月形
"""
x, y = datasets.make_moons(n_samples=50000, noise=0.3, random_state=42)
print(x.shape)
print(y.shape)
print(np.unique(y))
```

```python
plt.scatter(x[y == 0, 0], x[y == 0, 1])  #拿到特征中标签为0的第0列，拿到标签为0的第一列
plt.scatter(x[y == 1, 0], x[y == 1, 1]) #拿到特征中标签为1的第0列，拿到标签为1的第一列
plt.legend(['0', '1'])
plt.show()  #图形的点颜色是自动的
```

```python
#默认分割比例是75%和25%
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
#逻辑回归是搞分类的
log_clf = LogisticRegression()
log_clf.fit(x_train, y_train)
log_clf.score(x_test, y_test)
```

```python
from sklearn.svm import SVC

svm_clf = SVC() #支持向量机
svm_clf.fit(x_train, y_train)
svm_clf.score(x_test, y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt_clf = DecisionTreeClassifier()
dt_clf.fit(x_train, y_train)
dt_clf.score(x_test, y_test)
```

# 1 手动实现投票分类器

```python
#训练好模型，测试集做预测
#三个不同模型对同一批样本的二分类预测结果（每个值都是 0 或 1）
y_predict1 = log_clf.predict(x_test)
y_predict2 = svm_clf.predict(x_test)
y_predict3 = dt_clf.predict(x_test)
```

```python
#bagging，纯手动bagging
y_predict = np.array((y_predict1 + y_predict2 + y_predict3) >= 2, dtype='int')#判断票数是否达到多数（即至少两个模型预测为正），返回布尔值数组
y_predict[:10]
```

```python
y_test[:10]
```

```python
from sklearn.metrics import accuracy_score

#accuracy_score计算准确率的
accuracy_score(y_test, y_predict)
```

# 2 集成学习VotingClassifier(并行)

```python
from sklearn.ensemble import VotingClassifier

#hard模式就是少数服从多数
voting_clf = VotingClassifier(estimators=
                              [('log_clf', LogisticRegression()),
                               ('svm_clf', SVC()),
                               ('dt_clf', DecisionTreeClassifier())],
                              voting='hard')
```

```python
voting_clf.fit(x_train, y_train)
print(voting_clf.score(x_test, y_test))
voting_clf.predict(x_test)[:10]
```

更合理的投票，应该有权值
唱歌比赛的投票?(评委的投票权重与观众不一样)

```python
# hard和soft区别请看课件解释,训练时间久
voting_clf2 = VotingClassifier(estimators=
                               [('log_clf', LogisticRegression()),
                                ('svm_clf', SVC(probability=True)),#支持向量机中需要加入probability
                                ('dt_clf', DecisionTreeClassifier())],
                               voting='soft')
voting_clf2.fit(x_train, y_train)
voting_clf2.score(x_test, y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier  #用决策树集成上千模型
from sklearn.ensemble import BaggingClassifier

#估计器数目是5000呢？bootstrap是有放回抽样,max_samples不可以超过训练集的数目37500
"""
n_estimators=300：集成中基估计器的数量，即要训练的决策树数量为 300 棵。
max_samples=3000：每棵决策树从原始训练集中随机抽取的样本数量（有放回抽样，因为 bootstrap=True）。这里指定抽取 3000 个样本，如果原始训练集样本数少于 3000，则实际抽取量不超过训练集大小。
bootstrap=True：启用 Bootstrap 抽样（有放回），即每棵树的训练集是原始训练集的一个自助样本集，大小由 max_samples 指定。若设为 False，则使用整个训练集（但每棵树可能因特征抽样而不同）
"""
bagging_clf = BaggingClassifier(DecisionTreeClassifier(), 
                                n_estimators=300,
                                max_samples=3000, 
                                bootstrap=True)
bagging_clf.fit(x_train, y_train)
bagging_clf.score(x_test, y_test)
```

# 3 oob设计

```python
from sklearn.tree import DecisionTreeClassifier  #用决策树集成上千模型
from sklearn.ensemble import BaggingClassifier
#当去做有放回的抽样时，把没有取到的样本作为验证集，可以用oob_score来评估
bagging_clf2 = BaggingClassifier(DecisionTreeClassifier(), 
                                 n_estimators=500, 
                                 max_samples=1000,
                                 bootstrap=True,
                                 oob_score=True) #利用袋外样本（Out-of-Bag, OOB）进行模型评估
bagging_clf2.fit(x_train, y_train)
bagging_clf2.oob_score_  #拿没有取到的样本作为验证集
```

```python
bagging_clf2.score(x_test, y_test)
```

### njobs

```python
from sklearn.tree import DecisionTreeClassifier  #用决策树集成上千模型
from sklearn.ensemble import BaggingClassifier
```

```python
%%time
bagging_clf = BaggingClassifier(DecisionTreeClassifier(),
                                n_estimators=500, 
                                max_samples=2000,
                                bootstrap=True)
bagging_clf.fit(x_train, y_train)
bagging_clf.score(x_test, y_test)
#cpu time是实际cpu运行算法时间，wall time是墙上真实时间，就是real time（包含了进程启动时间，I/O等待时间等）
```

```python
%%time
#多核使用的时间更小, %%time要顶格来写
bagging_clf = BaggingClassifier(DecisionTreeClassifier(),
                                n_estimators=500, 
                                max_samples=2000,
                                bootstrap=True, 
                                n_jobs=-1)
bagging_clf.fit(x_train, y_train)
bagging_clf.score(x_test, y_test)
```

# 4 bootstrap_features  对特征进行采样

```python
#因为我们就两个特征，所以 max_features设置为 1 ,得到的分值较低是因为我们特征较少，采用随机特征不合适
#这个是只对特征进行采样
random_subspaces_clf = BaggingClassifier(DecisionTreeClassifier(),
                                         n_estimators=500, 
                                         max_samples=500, 
                                         oob_score=True,
                                         bootstrap=True, 
                                         n_jobs=-1,#使用所有可用的 CPU 核心并行训练 500 棵树，加速计算
                                         max_features=1, 
                                         bootstrap_features=True)
random_subspaces_clf.fit(x, y)
random_subspaces_clf.oob_score_
```

```python
# 这里是为了验证总的抽样的样本数不能大于训练集的数目
random_patches_clf = BaggingClassifier(DecisionTreeClassifier(),
                                       n_estimators=500, 
                                       max_samples=50001, 
                                       oob_score=True,
                                       bootstrap=True, 
                                       n_jobs=-1,
                                       max_features=1, 
                                       bootstrap_features=True)
random_patches_clf.fit(x, y)
random_patches_clf.oob_score_
#这个代码有另外一个名字就是随机森林
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
%%time
rc_clf = RandomForestClassifier(n_estimators=500, 
                                random_state=666,
                                oob_score=True, 
                                n_jobs=-1)
rc_clf.fit(x, y)
rc_clf.oob_score_
```

```python
%%time
#max_leaf_nodes 最大的叶子结点数,#限制每棵决策树的最大叶子节点数为 16。这是一种预剪枝策略，控制树的复杂度，防止单棵树过拟合。叶子节点数越少，树越简单，泛化能力可能更强（但可能欠拟合）。这个参数直接限制树的生长规模，与 max_depth 类似但更直接
rc_clf2 = RandomForestClassifier(n_estimators=500, 
                                 max_leaf_nodes=16,
                                 random_state=666,
                                 oob_score=True, 
                                 n_jobs=-1)#由于多核计算，因此wall time小于CPU times，因为每个核的计算时间会累加
rc_clf2.fit(x, y)
rc_clf2.oob_score_
```

# 5 Extra-Trees
ExtraTreesClassifier 是一种机器学习算法，属于基于决策树的集成方法家族。
它随机选择特征子集和数据样本来构建多个决策树，并将它们的预测聚合起来做出最终预测。
它可用于分类任务，并以处理嘈杂和高维数据的能力而闻名。

```python
from sklearn.ensemble import ExtraTreesClassifier

et_clf = ExtraTreesClassifier(n_estimators=500, 
                              oob_score=True,
                              bootstrap=True, 
                              n_jobs=-1)
et_clf.fit(x, y)
et_clf.oob_score_
```
